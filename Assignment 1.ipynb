```python
y = list(range(200))
y2 = [y**2 for y in range(200)]
plt.scatter(y, y2, c=y2, cmap=plt.cm.Greens, edgecolor='none', s=35)
plt.show()
```

```python
y = list(range(20))
y2 = [y**2 for y in range(20)]
y3 = [y**3 for y in range(20)]
plt.scatter(y, y2, c='blue', edgecolor = 'none', s = 20)
plt.scatter(y, y3, c='red', edgecolor = 'none', s = 20)
plt.axis([0,20, 0, 8000])
plt.show()
```
