```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
from sklearn import linear_model
```

```python
n = 10000
x = np.linspace(0.01, 1, n).reshape(-1, 1)
y = np.linspace(0.01, 1, n)  + np.random.rand(n) - .5


plt.scatter(x,y, color = 'lightyellow')
```

## 1. Create and fit a Linear Regression Model
## Calculate the Training error and Testing error using sklearn with a .50 split 
For error, use `mean_squared`, but if you want to experiment with other mean errors, please do!

# Linear regression model without `split`

```python
model0 = LinearRegression()
model0.fit(x, y)
model0.coef_, model0.intercept_
```

```python
print(x.shape, y.shape)
```

```python
plt.scatter(x, y, color = 'lightyellow')
plt.plot(x, np.dot(x, model0.coef_) + model0.intercept_, color = 'black', linewidth = '3')
```

# Linear regression with a 0.50 `split`

```python
from sklearn.model_selection import train_test_split
```

```python
x_train50, x_test50, y_train50, y_test50 = train_test_split(x, y, test_size = 0.5)
```

```python
print(x_train50.shape, y_train50.shape)
print(x_test50.shape, y_test50.shape)
```

```python
model_train_50 = LinearRegression()
model_train_50.fit(x_train50, y_train50)
model_train_50.coef_, model_train_50.intercept_
```

```python
plt.scatter(x_train50,y_train50, color = 'lightyellow')
plt.plot(x_train50, np.dot(x_train50, model_train_50.coef_) + model_train_50.intercept_,
         color = 'maroon', linewidth = '2')
```

```python
model_test_50 = LinearRegression()
model_test_50.fit(x_test50, y_test50)
model_test_50.coef_, model_test_50.intercept_
```

```python
plt.scatter(x_test50,y_test50, color = 'lightyellow')
plt.plot(x_test50, np.dot(x_test50, model_test_50.coef_) + model_test_50.intercept_,
         color = 'red', linewidth = '2')
```

```python
from sklearn.metrics import mean_squared_error
```

```python
train_50 = mean_squared_error(y_train50, np.dot(x_train50, model_train_50.coef_) + model_train_50.intercept_)
test_50 = mean_squared_error(y_test50, np.dot(x_test50, model_test_50.coef_) + model_test_50.intercept_)

print(train_50, test_50)
```
